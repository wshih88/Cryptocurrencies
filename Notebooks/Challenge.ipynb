```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import hvplot.pandas
```

```python
# Loading data
file_path = "../Resources/crypto_data.csv"
crypto_df = pd.read_csv(file_path)
crypto_df.head()
crypto_df.count()
```

```python
# Removing non-trading cryptos
df = crypto_df[crypto_df['IsTrading']==True]
df
```

```python
# Removing any blank Algorithm entries
dropna_df=df['Algorithm'].dropna()
dropna_df.count()
```

```python
# Removing 'IsTrading' column
drop_df = df.drop(columns = ['IsTrading'])
drop_df
```

```python
# Removing any null values
drop_df = drop_df.dropna()
drop_df
```

```python
# Removing TotalCoinsMined = 0
drop_df = drop_df[drop_df['TotalCoinsMined'] != 0]
drop_df
```

```python
# Removing TotalCoinSupply = 0
drop_df = drop_df[drop_df['TotalCoinSupply'] != '0']
drop_df
```

```python
drop_df.dtypes
```

```python
# Converting object to float
# https://kite.com/python/answers/how-to-convert-a-pandas-dataframe-column-of-strings-to-floats-in-python#:~:text=Use%20pandas.,the%20column%20values%20to%20floats.
drop_df['TotalCoinSupply'] = pd.to_numeric(drop_df['TotalCoinSupply'], downcast='float')
drop_df
```
